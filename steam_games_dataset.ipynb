**Check for free and paid games**

```python
free_games = (df["price"] == 0).sum()
paid_games = (df["price"] > 0).sum()

print(f"Number of free games: {free_games}")
print(f"Number of paid games: {paid_games}")
```

**Top 5 Games by Positive Reviews**

```python
top_games = df.sort_values(by="positive", ascending=False).head(5)

plt.figure(figsize=(10,5))
plt.bar(top_games["name"], top_games["positive"])

plt.xticks(rotation=90)
plt.title("Top 5 Games by Positive Reviews")
plt.xlabel("Game Name")
plt.ylabel("Positive Reviews")

plt.show()
```

**Game pricing distribution**

```python
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title("Game pricing distribution")
plt.xlabel("price")
plt.ylabel("Number of games")
plt.show()
```

**Relationship between price and number of positive reviews**

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='price', y='positive', data=df, alpha=0.6)
plt.title("The relationship between price and number of positive reviews")
plt.xlabel("price")
plt.ylabel("Number of positive reviews")
plt.show()
```

**Correlation heatmap**

```python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr = df[numeric_cols].corr()

plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Steam Games Dataset')
plt.show()
```

**Top 3 Games Analysis**

```python
df['owners_numeric'] = df['owners'].apply(lambda x: int(x.split('..')[0].replace(',', '')))
df['total_reviews'] = df['positive'] + df['negative']
df['positive_ratio'] = df['positive'] / df['total_reviews']
df['positive_ratio'] = df['positive_ratio'].fillna(0)
```

```python
top_owners = df.nlargest(3, 'owners_numeric')[['name', 'owners_numeric', 'positive', 'ccu']]
print("Top 3 by Owners:\n", top_owners)
```

```python
top_ccu = df.nlargest(3, 'ccu')[['name', 'ccu']]
print("\nTop 3 by Current Players:\n", top_ccu)
```

```python
top_ratio = df[df['total_reviews'] > 1000].nlargest(3, 'positive_ratio')[['name', 'positive_ratio', 'total_reviews']]
print("\nTop 3 by Positive Ratio:\n", top_ratio)
```

```python
plt.figure()
plt.bar(top_owners['name'], top_owners['owners_numeric'])
plt.title('Top 3 Games by Owners')
plt.xlabel('Game')
plt.ylabel('Number of Owners')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
plt.bar(top_ccu['name'], top_ccu['ccu'])
plt.title('Top 3 Games by Current Players (CCU)')
plt.xlabel('Game')
plt.ylabel('Current Players')
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure()
plt.bar(top_ratio['name'], top_ratio['positive_ratio'])
plt.title('Top 3 Games by Positive Ratio')
plt.xlabel('Game')
plt.ylabel('Positive Ratio')
plt.xticks(rotation=30)
plt.show()
```
